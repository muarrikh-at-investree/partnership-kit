```python
## Check Missing Value
def check_missing_value(df): 
    '''
     Fungsi get_missing_value
     Input berupa pandas dataframe
     Output/return menghasilkan tabel yang berisi kolom beserta jumlah dan presentase missing value pada dataset 
    '''

    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    missing_data = missing_data.loc[missing_data['Percent']>0]
    print(missing_data)
    if missing_data.shape[0]!=0:
        f, ax = plt.subplots(figsize=(5, 5))
        plt.xticks(rotation='90')
        sns.barplot(y=missing_data.index, x=missing_data['Percent'])
        plt.ylabel('Features', fontsize=15)
        plt.xlabel('Percent of missing values', fontsize=15)
        plt.xlim(0, 1)
        plt.title('Percent missing data by feature', fontsize=15)
        return missing_data
    else:
        print('Tidak ada missing value!!!')
```

```python
all_df = [df_supplier, df_po]

for data in all_df:
    check_missing_value(data)
```

Period

```python
##Supplier
try:
    df_supplier['SUPPLIER_PERIOD'] = None
    for i in range(0, len(df_supplier['REG_DATE'])):
        df_supplier['SUPPLIER_PERIOD'][i] = (refer_analysis_time.year - df_supplier['REG_DATE'][i].year) * 12 + (refer_analysis_time.month - df_supplier['REG_DATE'][i].month)

#     count_per_period = df_supplier['SUPPLIER_PERIOD'].value_counts().rename_axis('PERIOD').reset_index(name='JUMLAH')
#     print(count_per_period)

#     fig = plt.figure(figsize=(10,8))
#     ax = plt.subplot(111)
#     ax.bar(count_per_period.iloc[:, 0], count_per_period.iloc[:, 1])
    
    difference = round((max(df_supplier['SUPPLIER_PERIOD']) - min(df_supplier['SUPPLIER_PERIOD']))/5)
    group_1 = '< '+str(min(df_supplier['SUPPLIER_PERIOD'])+difference)+' Bulan'
    group_2 = str(min(df_supplier['SUPPLIER_PERIOD'])+difference)+' Bulan - '+str(min(df_supplier['SUPPLIER_PERIOD'])+2*difference-1)+' Bulan'
    group_3 = str(min(df_supplier['SUPPLIER_PERIOD'])+2*difference)+' Bulan - '+str(min(df_supplier['SUPPLIER_PERIOD'])+3*difference-1)+' Bulan'
    group_4 = str(min(df_supplier['SUPPLIER_PERIOD'])+3*difference)+' Bulan - '+str(min(df_supplier['SUPPLIER_PERIOD'])+4*difference-1)+' Bulan'
    group_5 = '> '+str(min(df_supplier['SUPPLIER_PERIOD'])+4*difference-1)+' Bulan'
    df_supplier['SUPPLIER_PERIOD_GROUP'] = None
    for i in range(0, len(df_supplier['SUPPLIER_PERIOD'])):
        if df_supplier['SUPPLIER_PERIOD'][i] < (min(df_supplier['SUPPLIER_PERIOD'])+difference):
            df_supplier['SUPPLIER_PERIOD_GROUP'][i] = group_1
        elif df_supplier['SUPPLIER_PERIOD'][i] < (min(df_supplier['SUPPLIER_PERIOD'])+2*difference):
            df_supplier['SUPPLIER_PERIOD_GROUP'][i] = group_2
        elif df_supplier['SUPPLIER_PERIOD'][i] < (min(df_supplier['SUPPLIER_PERIOD'])+3*difference):
            df_supplier['SUPPLIER_PERIOD_GROUP'][i] = group_3
        elif df_supplier['SUPPLIER_PERIOD'][i] < (min(df_supplier['SUPPLIER_PERIOD'])+4*difference):
            df_supplier['SUPPLIER_PERIOD_GROUP'][i] = group_4
        else:
            df_supplier['SUPPLIER_PERIOD_GROUP'][i] = group_5
    
    count_per_period = df_supplier['SUPPLIER_PERIOD_GROUP'].value_counts().rename_axis('PERIOD_GROUP').reset_index(name='JUMLAH')
    order = [group_1, group_2, group_3, group_4, group_5]
    count_per_period['PERIOD_GROUP'] = pd.Categorical(count_per_period['PERIOD_GROUP'], order)
    count_per_period.sort_values(by=['PERIOD_GROUP'], inplace=True)
    print(count_per_period)

    fig = plt.figure(figsize=(10,8))
    ax = plt.subplot(111)
    ax.barh(count_per_period.iloc[:, 0], count_per_period.iloc[:, 1])
    

except Exception as e:
    print(e)
```

Ceritified Documents (KYC)

```python
##Supplier
try:
    count_ktp = 0
    count_npwp = 0
    count_siup = 0
    count_tdp = 0
    for i in df_supplier['SUPPLIER_CERTIFIED_DOCUMENTS']:
        ktp = i.count('KTP')
        count_ktp = count_ktp+ktp
        npwp = i.count('NPWP')
        count_npwp = count_npwp+npwp
        siup = i.count('SIUP')
        count_siup = count_siup+siup
        tdp = i.count('TDP')
        count_tdp = count_tdp+tdp

    fig = plt.figure(figsize=(10,8))
    ax = plt.subplot(111)
    ax.bar(['KTP', 'NPWP', 'SIUP', 'TDP'], [count_ktp, count_npwp, count_siup, count_tdp])

    plt.show()
except Exception as e:
    print(e)
```
